```python
def word_rhyme(word,ss):
    tmp = []
    end = ss[word]
    for e in end:
        for key in ss.keys():
            if e in ss[key]:
                tmp.append(key)

    return set(tmp)
```

```python
from nltk.corpus import cmudict
# nltk.download('cmudict')

words = []

f = open("shakespeare.txt",'r')

for x in f:
    w = x.split(" ")
    for i in w:
        i = i.replace(',','').replace('\n','').replace(':','')
        if i != '' and i.isalpha():
            words.append(i.lower())


f.close()
words = set(words)

word_dict = cmudict.dict()
ss = {}
endings = []

for word in words:
    try:
        tmp = word_dict[word.lower()]
        mat = []
        for i in range(len(tmp)):
            w = tmp[i]
            for j in reversed(range(len(w))):
                if w[j][-1] == "1":
                    mat.append(w[j:])
                    break

        ss[word] = mat
    except:  # noqa: E722
        # print(f'{word} not in dict')
        pass
print(ss)
```
