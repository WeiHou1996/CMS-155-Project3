```python
def word_rhyme(ending,ss):
    tmp = []
    for key in ss.keys():
        if ending in ss[key]:
            tmp.append(key)

    return tmp
```

```python
from nltk.corpus import cmudict
# nltk.download('cmudict')

words = []

f = open("shakespeare.txt",'r')

for x in f:
    w = x.split(" ")
    for i in w:
        i = i.replace(',','').replace('\n','').replace(':','')
        if i != '' and i.isalpha():
            words.append(i.lower())


f.close()
words = set(words)

word_dict = cmudict.dict()
ss = {}
endings = []

for word in words:
    try:
        tmp = word_dict[word.lower()]
        mat = []
        for i in range(len(tmp)):
            if len(tmp[i]) > 5:
                mat.append(tmp[i][-4:])
                if tmp[i][-4:] not in endings:
                    endings.append(tmp[i][-3:]) 
            elif len(tmp[i])>3:
                mat.append(tmp[i][-3:])
                if tmp[i][-3:] not in endings:
                    endings.append(tmp[i][-3:])
            else:
                mat.append(tmp[i][-1:])
                if tmp[i][-2:] not in endings:
                    endings.append(tmp[i][-1:])
        ss[word] = mat
    except:  # noqa: E722
        # print(f'{word} not in dict')
        pass
print(ss)
```
